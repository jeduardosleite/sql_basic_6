### Query 2

```sql
SELECT DISTINCT
    transacoes.id_cliente, cliente.nome
FROM transacoes
INNER JOIN cliente
ON
    transacoes.id_cliente = cliente.id_cliente
ORDER BY id_cliente ASC;
```

````python
query_2
````

### Query 3

```sql
SELECT *
FROM cliente
CROSS JOIN transacoes;
```

### Query 4

```sql
SELECT *
FROM transacoes
LEFT JOIN cliente
ON cliente.id_cliente = transacoes.id_cliente;
```

### Query 5

```sql
SELECT *

RIGHT JOIN cliente
```
